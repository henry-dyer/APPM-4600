```python

import numpy as np
import matplotlib.pyplot as plt
```

```python

#HW1 Q1
def expanded(x):
     return 1*x**9 -18*x**8 +144*x**7 -672*x**6 +2016*x**5 -4032*x**4 +5376*x**3 -4608*x**2 +2304*x -512
```

```python

#HW1 Q1
def compact(x):
    return (x-2)**9
```

```python

x_values = np.arange(1.92, 2.081, 0.001)
y_values = expanded(x_values)

plt.plot(x_values, y_values, label='expanded function')
plt.title('P(x) via Coefficients')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python

y_values = compact(x_values)

plt.plot(x_values, y_values, label='expanded function')
plt.title('P(x) via Expression')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
